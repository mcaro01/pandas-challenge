```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# Files loaded from path
school_files = Path("C:\\Users\\mcaro\\OneDrive\\Desktop\\Homework\\pandas-challenge\\Resources\\schools_complete.csv")
student_files = Path("C:\\Users\\mcaro\\OneDrive\\Desktop\\Homework\\pandas-challenge\\Resources\\students_complete.csv")

#  School and Student Data File stored into Pandas DataFrames
school_list = pd.read_csv(school_files)
student_list = pd.read_csv(student_files)

# Combined the data into a single dataset.  
school_data = pd.merge(student_list, school_list, how="left", on=["school_name", "school_name"])
school_data.head()
```

# District Summary

```python
# Unique Schools total
unique_schools = school_data["school_name"].nunique()
unique_schools
```

```python
# Total number of Students
Total_students = school_data["student_name"].count()
Total_students
```

```python
# Total Budget
Total_budget = school_data["budget"].unique().sum()
Total_budget
```

```python
#Average (mean) math score 
Average_Math_Score = school_data["math_score"].mean()
Average_Math_Score
```

```python
# Average (mean) reading score
Average_reading_score = school_data["reading_score"].mean()
Average_reading_score
```

```python
# Percentage of students who passed math > = 70
Student_passed_math = school_data[(school_data["math_score"] >= 70)].count()["student_name"]
Percentage_students_passing_math = Student_passed_math / float(Total_students) * 100
Percentage_students_passing_math
```

```python
 # Percentage of students who passed reading > = 70 
Student_passed_reading = school_data[(school_data["reading_score"] >= 70)].count()["student_name"]
Percentage_students_passing_reading = Student_passed_reading / float(Total_students) * 100
Percentage_students_passing_reading
```

```python
# Percentage of students who passed math and reading  > = 70
Student_passed_math_reading = school_data[
    (school_data["math_score"] >= 70) & (school_data["reading_score"] >= 70)
].count()["student_name"]
percentage_overall_passing_rate = Student_passed_math_reading / Total_students * 100
percentage_overall_passing_rate
```

```python
#  high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools":[unique_schools], "Total Students": [Total_students],
                                 "Total Budget":[Total_budget],"Average Math Score":[Average_Math_Score],
                                 "Average Reading Score":[Average_reading_score],
                                 "% Passing Math":[Percentage_students_passing_math],
                                 "% Passing Reading": [Percentage_students_passing_reading],"% Overall Passing":
                                 [percentage_overall_passing_rate]
                                })

# Format
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

#  DataFrame Display
district_summary
```

# School Summary

```python
# Used the code provided for Schooltype
school_types = school_list.set_index(["school_name"])["type"]
```

```python
# Calculate the total student count
Total_student_count = school_data["school_name"].value_counts()
Total_student_count
```

```python
# Calculate the total school budget and per student budget
Total_school_budget = school_data.groupby(["school_name"]).mean()["budget"]
per_student_budget = Total_school_budget / Total_student_count
```

```python
# Calculate the average scores for math and reading
Aver_school_math = school_data.groupby(["school_name"]).mean()["math_score"]
Aver_school_reading = school_data.groupby(["school_name"]).mean()["reading_score"]
```

```python
#% of  number of schools with math scores of 70 or higher
Student_passed_math = school_data[(school_data["math_score"] >= 70)]
```

```python
# % of the number of schools with reading scores of 70 or higher
Student_passed_reading = school_data[(school_data["reading_score"] >= 70)]
```

```python
# Used the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
percentage_passing_math_and_reading = school_data[
    (school_data["reading_score"] >= 70) & (school_data["math_score"] >= 70)]
```

```python
#the provided code to calculate the passing rates
per_school_passing_math = Student_passed_math.groupby(["school_name"]).count()["student_name"] / Total_student_count * 100
per_school_passing_reading = Student_passed_reading.groupby(["school_name"]).count()["student_name"] / Total_student_count * 100
overall_passing_rate = percentage_passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / Total_student_count * 100 
```

```python
#  `per_school_summary` data frame  with columns for the calculations above.
per_school_summary = pd.DataFrame({"School Type":school_types, "Total Students": Total_student_count,
                                 "Total School Budget":Total_school_budget,"Per Student Budget": per_student_budget, 
                                  "Average Math Score":Aver_school_math,
                                 "Average Reading Score":Aver_school_reading,
                                 "% Passing Math":per_school_passing_math,
                                 "% Passing Reading": per_school_passing_reading,
                                  "% Overall Passing": overall_passing_rate
                                })
# Format
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
#per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

#  DataFrame Display
per_school_summary
```

# Highest-Performing Schools(by % Overall Passing) 

```python
#sorting % overall passing schools in descending order and display top 5 rows
top_schools =  per_school_summary.sort_values(by = "% Overall Passing", ascending=False)
top_schools.head()
```

# Scores by School spending

```python
# Created bins in which data is held
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"],
                                                             bins = spending_bins, labels = labels, include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({ "Average Math Score":spending_math_scores,
                                 "Average Reading Score":spending_reading_scores,
                                 "% Passing Math":spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                  "% Overall Passing": overall_passing_spending 
                                })

# Display results
spending_summary
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({"Average Math Score":size_math_scores,
                                 "Average Reading Score":size_reading_scores,
                                 "% Passing Math":size_passing_math,
                                 "% Passing Reading": size_passing_reading,
                                  "% Overall Passing": size_overall_passing 
                                })


# Display results
size_summary
```
